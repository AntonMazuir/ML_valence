```python
import os
import json
import pandas as pd
from pathlib import Path

# On remonte d'un niveau pour atteindre la racine du projet
root_dir = Path(os.getcwd()).parent
data_dir = root_dir / "data" / "raw"

# On cherche le fichier le plus r√©cent
list_of_files = list(data_dir.glob('*.json'))
if list_of_files:
    latest_file = max(list_of_files, key=os.path.getctime)
    print(f"üìÇ Chargement de : {latest_file.name}")

    with open(latest_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    df = pd.DataFrame(data['elementList'])
    print(f"‚úÖ DataFrame cr√©√© : {len(df)} annonces.")
else:
    print("‚ùå Aucun fichier trouv√©. V√©rifie le chemin !")
```
